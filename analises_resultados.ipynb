```python
import pandas as pd
from src.agent import validate_question, generate_sql
from src.database import execute_read_query
from src.utils import anonimizar_dados, gerar_grafico
from src.agent import validate_question, generate_sql, interpretar_dados
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
def perguntar_ao_agente(pergunta):
    print(f"👤 Pergunta: {pergunta}\n")
    
    validacao = validate_question(pergunta)
    if not validacao.is_valid:
        print(f"🛡️ Guardrail: {validacao.reason}")
        return
        
    resposta = generate_sql(pergunta)
    print("💻 SQL Gerado:")
    print(resposta.sql_query)
    print(f"\n🧠 Explicação: {resposta.explanation}\n")
    
    df = execute_read_query(resposta.sql_query)
    
    if isinstance(df, str):
        print(f"❌ Erro SQL: {df}")
        return
        
    df_seguro = anonimizar_dados(df)
    
    display(df_seguro)
    
    print("\n🧐 Análise do Especialista:")
    analise = interpretar_dados(pergunta, df_seguro.to_markdown(index=False))
    print(analise)
    print("\n" + "-"*50 + "\n")
    
    if resposta.visualizacao_recomendada != "nenhum" and resposta.visualizacao_recomendada != "tabela":
        gerar_grafico(df_seguro, resposta.visualizacao_recomendada)
```
